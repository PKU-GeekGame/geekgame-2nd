```python
import sys, random
import numpy as np
import hashlib
import re
```

```python
letters='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
filename='src/crypt2.txt'

with open(filename) as f:
    plain_text=f.read()
```

```python
code_pair=[]

for i in letters:
    il=i.lower()
    md5=hashlib.md5(il.encode('utf-8')).hexdigest()
    print(il,md5)
    pa=re.sub("\D",r'\\D',md5)
    try:
        start=re.search(pa,plain_text).start()
    except:
        continue

    counter=0
    a_list=[]
    a_class={i.lower():0 for i in letters}

    for i in range(len(plain_text)):
        if i==start:
            print(' ',plain_text[i:i+32],counter)
            encode=plain_text[i:i+32]
            break
        l=plain_text[i]
        if l not in letters and l not in letters.lower():
            continue
        else:
            a_class[l.lower()]=a_class[l.lower()]+1
            
            counter=counter+1

    base=counter
    orgin= md5
    encode=encode

    for i in range(len(orgin)):
        p1=orgin[i]
        p2=encode[i]
        if p1 in letters or p1 in letters.lower():
            #print(p1+'-'+p2,base)
            code_pair.append([p1,p2,np.mod(base,26)])
            base=base+1
```

```python
code_pair.sort(key=lambda x:x[2])
code_pair.sort(key=lambda x:x[0])
code_pair
```

```python
n=len(code_pair)
for i in range(1,n):
    n0=code_pair[i][0]
    n1=code_pair[i][1]
    n2=code_pair[i][2]
    
    pre0=code_pair[i-1][0]
    pre1=code_pair[i-1][1]
    pre2=code_pair[i-1][2]

    if n0==pre0 and n1!=pre1:
        code_pair.append([pre1,n1,n2-pre2])
        if n2-pre2<0:
            print(i)

code_pair.sort(key=lambda x:x[2])
code_pair.sort(key=lambda x:x[0])
```

```python
keymap={}
for pair in code_pair:
    if pair[2]==1:
        keymap[pair[1].upper()]=pair[0].upper()
```
